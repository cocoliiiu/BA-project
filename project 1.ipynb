```python
import pandas as pd
```

```python
df=pd.read_pickle(r"shared/Motor_Vehicle_Collisions_-_Crashes.pkl")
```

```python
df
```

```python
#1
# Filter data after year 2012 
df['CRASH DATE_CRASH TIME']=pd.to_datetime(df['CRASH DATE_CRASH TIME'])
date_df=df[df['CRASH DATE_CRASH TIME'].dt.year>=2012]
# count the total number of crashes by borough
crash_counts = date_df['BOROUGH'].value_counts()
crash_counts
```

```python
#2
#add the population
BRONX_pop=1446788 
BROOKLYN_pop=2648452 
MAN_pop= 1638281 
QUEENS_pop=2330295
STAT_pop=487155 
#get the number of region crashes
Bronx_df = crash_counts['BRONX']
Brook_df=crash_counts['BROOKLYN']
Man_df=crash_counts['MANHATTAN']
Que_df=crash_counts['QUEENS']
Sta_df=crash_counts['STATEN ISLAND']
#in every 100,000 people
Bronx_crash=Bronx_df/BRONX_pop*100000
Brook_crash=Brook_df/BROOKLYN_pop*100000
Man_crash=Man_df/MAN_pop*100000
Que_crash=Que_df/QUEENS_pop*100000
Sta_crash=Sta_df/STAT_pop*100000
Bronx_crash,Brook_crash,Man_crash,Que_crash,Sta_crash
```

```python
#3
#select the unspecific values
unspecific_drop=df[df['CONTRIBUTING FACTOR VEHICLE 1']=='Unspecified'].index
#drop the unspecific ones
df_dropped=df.drop(unspecific_drop)
df_dropped['CONTRIBUTING FACTOR VEHICLE 1'].value_counts(normalize=True)
```

```python
#4
df_dropped['CONTRIBUTING FACTOR VEHICLE 1'].value_counts(normalize=True).cumsum()
```

```python
#5
people_died_overtwo= date_df[date_df['NUMBER OF PERSONS KILLED']>=2]
accident_count=len(people_died_overtwo)
accident_count
```

```python
#6
#calculate the accidents with more than one people died
people_died_overone=date_df[date_df['NUMBER OF PERSONS KILLED']>=1]
accident_count_1=len(people_died_overone)
#calculate the total accidents
total_accidents=len(date_df)
#calculate the accidents have resulted in at least one person dead for every 1000 people
every_1000=accident_count_1/total_accidents*1000
every_1000
```

```python
#7
#Calculate the accidents without cross street names recorded
no_street_name=df['CROSS STREET NAME'].isna().sum
# Calculate the proportion 
no_street_name_prop=no_street_name/total_accidents
no_street_name_prop
```

```python
#8
#group by the the combinations
vehicle_combination = df.groupby(['VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2']).size()
#sort the counts in descending order
most_common = vehicle_combination.sort_values(ascending=False)
most_common
```

```python
#9
#number of accidents related to alcohol in contributing factor vehicle 1
alcohol_df=df[df['CONTRIBUTING FACTOR VEHICLE 1']=='Alcohol Involvement']
#number of accidents made fatalities with alcohol
Acc_df=alcohol_df[alcohol_df['NUMBER OF PERSONS KILLED'] > 0]
len(Acc_df)/len(alcohol_df)
```

```python
#10
#Calculate the accidents happened in the rush time
crash_in_rush=len(df[(df['CRASH DATE_CRASH TIME'].dt.hour >= 16) & (df['CRASH DATE_CRASH TIME'].dt.hour < 19)])
#Calculate the proportion
crash_in_rush/total_accidents
```

```python
#11
#filter the value with motorcycle in V1V2
V1_V2=df[(df['VEHICLE TYPE CODE 1'].str.contains('MOTORCYCLE', na=False)) |
          (df['VEHICLE TYPE CODE 2'].str.contains('MOTORCYCLE', na=False))]
#calculate the number of accidents involves injuries caused by motor in V1V2
motor_injured=V1_V2[V1_V2['NUMBER OF PERSONS INJURED'] > 0]
#calulate the number of accidents that do not involves dead casued by motor in V1V2
motor_no_fata=V1_V2[V1_V2['NUMBER OF PERSONS KILLED'] == 0]
#calculate the proportion
len(motor_no_fata)/len(motor_injured)
```

```python
#12
#filter the values with bicycle in V1V2
bicycle_crashes = df[(df['VEHICLE TYPE CODE 1'].str.contains('BICYCLE', na=False)) |
                     (df['VEHICLE TYPE CODE 2'].str.contains('BICYCLE', na=False))]
#count the number
num_bicycle_crashes = len(bicycle_crashes)
num_bicycle_crashes
```
